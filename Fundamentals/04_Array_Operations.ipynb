## 🧩 What Are Array Operations?

Array operations allow you to perform calculations directly on NumPy arrays without writing loops

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Without vectorization — the slow way</span></div>

````python
import numpy as np
array= np.arange(4)
print(array)
result= []
for i in array:
    result.append(i+10)

print(result)    
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ With vectorization — the NumPy way</span></div>

````python
result = array+10
print(result)
````

This is called vectorized computation.<br>Without array operations, AI would be much slower.

<div style="border:2px solid #8B5CF6; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#7C3AED; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">🤖</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">AI Insight — why vectorization matters</span></div>
<div style="background:#F5F3FF; padding:14px 18px; color:#3B0764; font-size:0.97em; line-height:1.6;">
Vectorized operations run in optimized C loops under the hood, avoiding Python's per-element interpreter overhead. This is the difference between training on millions of rows in seconds vs. minutes.
</div>
</div>

Wrong:

````python
(arr > 10) and (arr < 40)
````

Correct: 

````python
(arr > 10) & (arr < 40)
````

<div style="border:2px solid #F59E0B; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#D97706; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">⚠️</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">Warning — and/or vs &/|</span></div>
<div style="background:#FFFBEB; padding:14px 18px; color:#78350F; font-size:0.97em; line-height:1.6;">
Python's <code>and</code>/<code>or</code> expect a single Boolean, but a NumPy comparison returns an <b>array</b> of Booleans — raising a <code>ValueError: truth value is ambiguous</code>. Always use the bitwise operators <code>&amp;</code> and <code>|</code> on arrays instead.
</div>
</div>

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ ❌ Forgetting Parentheses</span></div>

Wrong: 

````python
arr > 10 & arr < 40
````

Correct: 

````python
(arr > 10) & (arr < 40)
````

The parentheses ensure each comparison is evaluated before applying the logical operator.

<div style="border:2px solid #2563EB; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#1D4ED8; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">💡</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">Tip — operator precedence trap</span></div>
<div style="background:#EFF6FF; padding:14px 18px; color:#1E3A5F; font-size:0.97em; line-height:1.6;">
Bitwise operators (<code>&amp;</code>, <code>|</code>) have <b>higher precedence</b> than comparison operators in Python. Without parentheses, <code>arr &gt; 10 &amp; arr &lt; 40</code> is evaluated as <code>arr &gt; (10 &amp; arr) &lt; 40</code> — always wrap each comparison: <code>(arr &gt; 10) &amp; (arr &lt; 40)</code>.
</div>
</div>

## 🎤 Interview Questions

---

### 🟢 Beginner

**Q1.** What is a vectorized operation, and why is it faster than a Python loop?
<details><summary>▶ Answer</summary>
A vectorized operation applies a calculation to every array element simultaneously using optimized, compiled C code under the hood — avoiding the overhead of Python's per-iteration interpreter loop.
</details>

**Q2.** What type of array does a comparison operation like <code>arr > 20</code> return?
<details><summary>▶ Answer</summary>
A Boolean array of the same shape, with <code>True</code>/<code>False</code> at each position based on the comparison.
</details>

**Q3.** What does <code>array / 2</code> return if <code>array</code> is an integer array?
<details><summary>▶ Answer</summary>
A <b>float</b> array — division always promotes the result to floating point, even if inputs are integers.
</details>

---

### 🟡 Intermediate

**Q4.** Why does <code>(arr > 10) and (arr < 40)</code> raise an error?
<details><summary>▶ Answer</summary>
Python's <code>and</code> requires a single Boolean value, but each side is an <b>array</b> of Booleans — NumPy can't collapse an array to one truth value, so it raises <code>ValueError: truth value is ambiguous</code>. Use <code>&amp;</code> instead.
</details>

**Q5.** What's the difference between <code>*</code> and <code>@</code> for two 2-D arrays?
<details><summary>▶ Answer</summary>
<code>*</code> is element-wise multiplication (Hadamard product) — shapes must match or broadcast. <code>@</code> (or <code>np.matmul()</code>) is true matrix multiplication, following linear algebra rules.
</details>

**Q6.** Why must comparisons be wrapped in parentheses before combining with <code>&amp;</code> or <code>|</code>?
<details><summary>▶ Answer</summary>
Bitwise operators have higher precedence than comparison operators in Python, so without parentheses the expression is grouped incorrectly, often producing wrong results or an error.
</details>

---

### 🔴 Advanced

**Q7.** How would you use Boolean masking to replace all negative values in an array with 0?
<details><summary>▶ Answer</summary>
<code>arr[arr &lt; 0] = 0</code> — the Boolean mask selects only the negative positions, and assignment updates them in place.
</details>

**Q8.** Why does floor division (<code>//</code>) discard decimals instead of rounding?
<details><summary>▶ Answer</summary>
Floor division always rounds <b>toward negative infinity</b> (truncates the fractional part downward), which differs from standard rounding — important for indices or counts where you always want a whole number ≤ the true quotient.
</details>

**Q9.** What happens when you apply <code>~</code> (NOT) to a Boolean array — does it use Python's <code>not</code> semantics?
<details><summary>▶ Answer</summary>
No — <code>~</code> is the bitwise NOT, which NumPy overloads to flip each Boolean element-wise. Python's <code>not</code> only works on a single value and would error on an array, just like <code>and</code>/<code>or</code>.
</details>

## 📌 Summary

### Key Takeaways

- **Vectorized operations** apply to every array element at once — no explicit loop needed, and they run in optimized C code.
- **Scalar operations** (`+`, `-`, `*`, `/`, `//`, `%`, `**`) broadcast the scalar across every element.
- **Array-to-array operations** apply element-wise when shapes match (or are broadcast-compatible).
- **`*` is element-wise multiplication** — use `@` / `np.matmul()` for true matrix multiplication.
- **Comparison operators** (`>`, `<`, `==`, `!=`, `>=`, `<=`) return Boolean arrays — the basis of masking/filtering.
- **Logical operators must be `&`, `|`, `~`** on arrays — Python's `and`/`or`/`not` will raise errors.
- **Always parenthesize** each comparison before combining with `&`/`|` — bitwise operators bind tighter than comparisons.

### Function Quick Reference

| Category | Operators | Returns |
|---|---|---|
| Scalar arithmetic | `+ - * / // % **` | Array (broadcast) |
| Array-to-array | `+ - * / // %` | Array (element-wise) |
| Comparison | `> < == != >= <=` | Boolean array |
| Logical | `& \| ~` | Boolean array |
