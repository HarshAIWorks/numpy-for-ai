> **The universal constructor.** Converts any sequence-like object — list, tuple, range, another array — into a NumPy `ndarray`.

**Syntax:**
```python
np.array(object, dtype=None, copy=True, order='K', subok=False, ndmin=0)
```

| Parameter | Description | Default |
|-----------|-------------|---------|
| `object` | Input data (list, tuple, array…) | Required |
| `dtype` | Override the data type | Auto-detected |
| `copy` | Create a new array or share memory | `True` |
| `ndmin` | Minimum number of dimensions | `0` |

````python
# 1-D
a =np.array([1,2,3])
a
````

````python
# 2-D and 3-D
b= np.array([[1,2,3],[4,5,6]])

c=np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print("2-D array: ",b)
print("3-D array: ",c) #3-D creates like- make list of list and each list have another lists 
````

### 6️⃣ Upcasting (Type Promotion)

NumPy enforces a **single dtype** per array. When you mix types, it promotes to the most capable type:

`bool` → `int` → `float` → `complex` → `str` → `object`

 Int + float ---> float items and type

````python
arra=np.array([1,2.5,3])
print(arra,arra.dtype)
````

Int + Float + Complex --> Complex items and Datatypes

````python
arrb= np.array([1,2,3.5,4.0,2+3j])
print(arrb,arrb.dtype)
````

Boolean with Numbers - Number Items and number type

True -- 1
False -- 0


````python
arrc= np.array([1,2,True])
print(arrc,arrc.dtype)
````

### 7️⃣ Dictionary and Set — Important AI Engineer Notes

Dictionaries and sets passed directly to `np.array()` are treated as a **single Python object** (0-D array, `dtype=object`). You must iterate their keys/values/items explicitly.

<div style="border:2px solid #8B5CF6; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#7C3AED; padding:8px 16px; display:flex; align-items:center; gap:8px;">
<span style="font-size:1.1em;">🤖</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">AI Insight</span>
</div>
<div style="background:#F5F3FF; padding:14px 18px; color:#3B0764; font-size:0.97em; line-height:1.6;">
In ML pipelines, feature dictionaries (<code>{feature: value}</code>) are common. Always extract values with <code>list(d.values())</code> before passing to NumPy — never pass the dict directly.
</div>
</div>

Dictionary inside - 0-D array

````python
dictionary = np.array({
    'Name': 'Harsh',
    'class': 10,
    'age' : 16
})
print(dictionary) #NumPy treats the entire dictionary as a single Python object, not as multiple elements.
                  #Shape: 0-D
````

Creating an array from dictionary keys

````python
dic = {
    'Name': 'Harsh',
    'class': 10,
    'age' : 16
}
keys= np.array(list(dic.keys())) #let create a list of items comes from keys- .keys() give all keys
print(keys)
````

Creating an array from dictionary values

````python
values= np.array(list(dic.values()))
print(values)
````

Creating an array from dictionary items

````python
items= np.array(list(dic.items()))
print(items)
````

### 8️⃣ `None` vs `np.nan`

| | `None` | `np.nan` |
|--|--------|----------|
| Python type | `NoneType` object | `float` |
| NumPy dtype when mixed with numbers | forces `object` | stays `float64` |
| Math operations | raises `TypeError` | propagates `nan` |
| Best used for | non-numeric missing data | numeric missing data in ML |

<div style="border:2px solid #8B5CF6; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#7C3AED; padding:8px 16px; display:flex; align-items:center; gap:8px;">
<span style="font-size:1.1em;">🤖</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">AI Insight — Missing Data in ML</span>
</div>
<div style="background:#F5F3FF; padding:14px 18px; color:#3B0764; font-size:0.97em; line-height:1.6;">
Always prefer <code>np.nan</code> over <code>None</code> for missing numeric values. pandas and scikit-learn detect and handle <code>np.nan</code> efficiently. Using <code>None</code> forces <code>dtype=object</code>, disabling vectorised computation.
</div>
</div>

````python
arr = np.array([1, 2, None])

print(arr)
print(arr.dtype)
````

````python
arr = np.array([1, 2, np.nan])

print(arr)
print(arr.dtype) # float output 
````

### 9️⃣ Copying an Existing NumPy Array

By default, `np.array(existing_array)` makes a **deep copy** — modifying the original will not affect the copy.  
Setting `copy=False` allows both variables to share the same memory block (a view).

````python
old= np.array(["hello",1,5.0])
new= np.array(old)
print(new) #upcasting: array prioroty- array items
````

Are both same objects?

````python
old is new
````

Any change in old will not affect new one

````python
old[0]=100
print(old)
print(new)
````

Copy=False

````python
#because copy = true by default parameter 
c=np.array([1,20,"good"])
d=np.array(c, copy=False)
c is d # now chnages also happen in another one
````

> **NumPy's version of Python's built-in `range()`** — returns an `ndarray` and supports float steps.

**Syntax:**
```python
np.arange([start,] stop[, step], dtype=None)
```

| Parameter | Description | Default |
|-----------|-------------|---------|
| `start` | Starting value (included) | `0` |
| `stop` | Ending value (**excluded**) | Required |
| `step` | Spacing between values (cannot be 0) | `1` |
| `dtype` | Override data type | Auto |

Creates a NumPy array with evenly spaced values using a fixed step.

## 🎤 Interview Questions

---

### 🟢 Beginner

**Q1.** What is the difference between `np.array([1, 2, 3])` and `np.arange(3)`?  
<details><summary>▶ Answer</summary>
<code>np.array([1, 2, 3])</code> creates an array from an explicit list: <code>[1, 2, 3]</code>.  
<code>np.arange(3)</code> generates a range starting from 0: <code>[0, 1, 2]</code>.  
They produce <b>different values</b>; only the shape is the same.
</details>

**Q2.** What dtype does NumPy assign to `np.array(['apple', 'mango'])`? Why?  
<details><summary>▶ Answer</summary>
<code>&lt;U5</code> — Unicode string of max length 5 (the length of <code>'apple'</code>). NumPy sets the width to the longest string in the array.
</details>

**Q3.** What is the output of `np.array([]).dtype`?  
<details><summary>▶ Answer</summary>
<code>float64</code> — empty arrays default to float64.
</details>

---

### 🟡 Intermediate

**Q4.** What is upcasting in NumPy? Give an example.  
<details><summary>▶ Answer</summary>
NumPy promotes all elements to the most general dtype when types are mixed.  
Example: <code>np.array([1, 2.5, 3])</code> → all elements become <code>float64</code> because 2.5 requires float.  
Promotion order: <code>bool → int → float → complex → str → object</code>
</details>

**Q5.** What is the difference between `np.zeros()` and `np.empty()`?  
<details><summary>▶ Answer</summary>
<code>np.zeros()</code> initialises every element to 0 (safe to read immediately).  
<code>np.empty()</code> allocates memory without writing any values — faster, but unsafe to read before filling.
</details>

**Q6.** When would you use `np.linspace()` instead of `np.arange()`?  
<details><summary>▶ Answer</summary>
When you need an exact count of points between two values.  
<code>np.arange()</code> can produce unexpected element counts with float steps due to floating-point precision.  
<code>np.linspace()</code> guarantees exactly <code>num</code> elements.
</details>

---

### 🔴 Advanced

**Q7.** What is the difference between `np.random.shuffle()` and `np.random.permutation()`?  
<details><summary>▶ Answer</summary>
<code>shuffle()</code> modifies the array <b>in-place</b> and returns <code>None</code>.  
<code>permutation()</code> returns a <b>new shuffled copy</b> and leaves the original unchanged.
</details>

**Q8.** Why does `np.array({'a': 1, 'b': 2})` produce a 0-D object array instead of a 1-D array?  
<details><summary>▶ Answer</summary>
NumPy does not iterate over dictionaries (or sets). It wraps the entire dict as a single Python object, producing a 0-D array with <code>dtype=object</code>.  
Use <code>list(d.values())</code> to get a proper 1-D numeric array.
</details>

**Q9.** What is `np.logspace(0, 3, 4)` equivalent to, and when would you use it?  
<details><summary>▶ Answer</summary>
Equivalent to <code>10 ** np.linspace(0, 3, 4)</code> → <code>[1., 10., 100., 1000.]</code>  
Used for log-scale hyperparameter grids — e.g. sweeping learning rates from 1e-4 to 1e-1.
</details>

## 📌 Summary

### Key Takeaways

- **`np.array()`** is the universal constructor — it accepts lists, tuples, arrays, and scalars. It always makes a deep copy by default.
- **NumPy enforces a single dtype** per array. Mixing types triggers automatic upcasting: `bool → int → float → complex → str → object`.
- **`np.arange()`** works like `range()` but returns an ndarray and supports float steps — use with care due to floating-point precision.
- **`np.linspace()`** is the safe choice when you need an exact count of evenly spaced float values.
- **`np.logspace()`** creates logarithmically spaced sequences — essential for hyperparameter grids.
- **`np.zeros()`, `np.ones()`, `np.full()`** create pre-filled arrays for initialising weights and biases.
- **`np.empty()`** is faster but dangerous — never read before writing every element.
- **`np.eye()`** supports rectangular and offset diagonals; `np.identity()` is its simpler square-only version.
- **`np.random`** functions power stochastic ML workflows — always `seed()` for reproducibility.
