NumPy is built for speed and low memory usage. Understanding the copy/view distinction lets you avoid silent data corruption (modifying a view you thought was a copy) and unnecessary memory overhead (copying large arrays you could have shared).
</div>
</div>

**Copy** — two independent arrays, each with their own memory:
```
arr        copy_arr
 │              │
 ▼              ▼
[10 20 30]  [10 20 30]
```
Changing one does **not** affect the other.

**View** — two Python objects pointing at the same underlying data:
```
           ┌───────────────┐
arr ──────►│  [10 20 30]   │
view_arr ─►│               │
           └───────────────┘
```
Changing one **does** affect the other.

<a id="fn-1"></a>

<span style="background:#2A2F47; color:#FFFFFF; font-weight:800; font-size:1.05em; padding:4px 12px; border-radius:6px;">1</span>
<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">View — Shared Memory</code>
</div>

A **view** is a new NumPy array object that shares the exact same data buffer as the original. Created with `arr.view()` or produced automatically by slicing.

````python
import numpy as np 
arr = np.array([10,20,30,40]) 
view_arr = arr.view() 
view_arr[0]=111
print(view_arr)
arr is view_arr
print(id(arr))
id(view_arr)
arr
````

```
Python Object 1 (arr)
        │
        ▼
   ┌──────────────┐
   │ Data Buffer  │  ← same memory block
   │ 10 20 30 40  │
   └──────────────┘
        ▲
        │
Python Object 2 (view_arr)
```
`arr` and `view_arr` are different Python objects (`id()` differs), but they share the same underlying data buffer.

<div style="border:2px solid #F59E0B; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#D97706; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">⚠️</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">Common Misconception</span></div>
<div style="background:#FFFBEB; padding:14px 18px; color:#78350F; font-size:0.97em; line-height:1.6;">
<code>id(arr) != id(view_arr)</code> — they are different Python objects. <b>Don't</b> use <code>id()</code> to check memory sharing; it compares object identity, not data buffer identity.
</div>
</div>

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:16px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Proving Shared Memory</span></div>

<div style="margin:10px 0 4px 0; font-family:'Segoe UI',sans-serif;"><span style="color:#64748B; font-weight:600; font-size:0.88em; font-style:italic;">— Method 1 (Recommended) — .base</span></div>

````python
print(view_arr.base is arr)
````

`True` confirms `view_arr` is backed by `arr`'s data.

<div style="margin:10px 0 4px 0; font-family:'Segoe UI',sans-serif;"><span style="color:#64748B; font-weight:600; font-size:0.88em; font-style:italic;">— Method 2 (Definitive) — .ctypes.data buffer address</span></div>

````python
print(arr.ctypes.data)
print(view_arr.ctypes.data)
````

```
id()          — compares Python objects        (NOT reliable for memory check)
.ctypes.data  — compares the actual data buffer address  
.base         — tells you whether the array shares memory 
```

<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">Copy — Independent Memory</code>
</div>

A **copy** creates a brand-new array with its own allocated data buffer. Changes to the copy never affect the original, and vice versa.

````python
copy_arr = arr.copy()
copy_arr
````

```
arr           copy_arr
 │                 │
 ▼                 ▼
[10 20 500 40]  [10 200 30 40]
```
Two completely separate memory locations — modifying either has no effect on the other.

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:16px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Confirming with .base</span></div>

````python
print(copy_arr.base)
````

`None` means `copy_arr` owns its own data and shares memory with nothing.

<span style="background:#2A2F47; color:#FFFFFF; font-weight:800; font-size:1.05em; padding:4px 12px; border-radius:6px;">3</span>
<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">Operations: View vs Copy ⭐⭐⭐⭐⭐</code>
</div>

| Operation | Returns |
|---|---|
| `arr.view()` | ✅ View |
| Basic slicing `arr[:]` | ✅ View |
| `reshape()` / `ravel()` | ✅ View *(usually — copy if layout forces it)* |
| `transpose()` / `.T` | ✅ View |
| `arr.copy()` | ✅ Copy |
| Fancy indexing `arr[[0,2]]` | ✅ Copy |
| Boolean masking `arr[arr > 5]` | ✅ Copy |
| Arithmetic `arr + 5`, `arr * 2` | ✅ Copy |
| `flatten()` | ✅ Copy |

<div style="border:2px solid #2563EB; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#1D4ED8; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">💡</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">Quick Rule</span></div>
<div style="background:#EFF6FF; padding:14px 18px; color:#1E3A5F; font-size:0.97em; line-height:1.6;">
<b>Slicing → View.</b> Everything that jumps around non-contiguously (fancy indexing, boolean masking) → Copy. Arithmetic → Copy (new values).
</div>
</div>

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:16px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Example 1 — Slicing Returns a View</span></div>

````python
arr = np.array([10,20,30,40,50])
slc=arr[:4]
slc[0]=101
arr
````

Modifying `slc[0]` changed `arr` too — proof that slicing is a view. This is by design: it lets you efficiently edit a *window* into the original.

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:16px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Example 2 — Fancy Indexing Returns a Copy</span></div>

````python
arr = np.array([10,20,30,40])
fnc=arr[[1,3]]
fnc[1]=100
print(arr)
fnc
````

`arr` is unchanged — fancy indexing always returns an independent copy.

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:16px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Example 3 — Boolean Masking Returns a Copy</span></div>

````python
fil = arr[arr > 20]
fil[0] = 999
print(arr)
````

Same result: `arr` is unchanged because boolean masking returns a copy.

## 🎤 Interview Questions

---

### 🟢 Beginner

**Q1.** What is the key difference between a view and a copy?
<details><summary>▶ Answer</summary>
A view shares the underlying data buffer with the original array — modifying one modifies the other. A copy has its own independent buffer — changes are isolated.
</details>

**Q2.** How do you create a view and a copy of an array?
<details><summary>▶ Answer</summary>
View: <code>arr.view()</code> or a basic slice like <code>arr[1:4]</code>. Copy: <code>arr.copy()</code>.
</details>

**Q3.** What does `copy_arr.base` return for a copy, and what does it return for a view?
<details><summary>▶ Answer</summary>
For a copy: <code>None</code> — it owns its own data. For a view: the original array it borrows data from.
</details>

---

### 🟡 Intermediate

**Q4.** Why can't you use `id()` to check whether two arrays share memory?
<details><summary>▶ Answer</summary>
<code>id()</code> returns the identity of the Python object, not the address of the underlying data buffer. A view is a <i>different</i> Python object (different <code>id</code>) that happens to point at the same raw memory.
</details>

**Q5.** Does slicing a 2D array with `arr[:2, :]` return a view or a copy?
<details><summary>▶ Answer</summary>
A view — basic slicing (using <code>:</code> notation) always returns a view in NumPy.
</details>

**Q6.** Does `arr.flatten()` return a view or a copy? How about `arr.ravel()`?
<details><summary>▶ Answer</summary>
<code>flatten()</code> always returns a copy. <code>ravel()</code> returns a view whenever possible; it falls back to a copy only if the array's memory layout prevents sharing.
</details>

---

### 🔴 Advanced

**Q7.** You slice an array and pass the slice to a function that modifies it. The original array changes unexpectedly. What happened and how do you prevent it?
<details><summary>▶ Answer</summary>
The slice is a view — in-place modifications propagate back to the original. To prevent it, explicitly pass <code>arr[1:4].copy()</code> into the function.
</details>

**Q8.** Why does fancy indexing always return a copy instead of a view?
<details><summary>▶ Answer</summary>
The selected elements may be scattered non-contiguously in memory (e.g. <code>arr[[0,5,3]]</code>). NumPy can only create a view over contiguous or regularly-strided memory; arbitrary index lists can't be described that way, so a copy is the only option.
</details>

**Q9.** How can you determine at runtime whether `b = some_operation(arr)` produced a view or a copy?
<details><summary>▶ Answer</summary>
Check <code>b.base is arr</code> — if <code>True</code>, it's a view sharing <code>arr</code>'s data. If <code>False</code> (i.e. <code>b.base is None</code> or another object), it's a copy or a view of something else.
</details>

## 📌 Summary

- A **view** shares the original array's data buffer — fast and memory-efficient, but in-place changes propagate.
- A **copy** owns independent memory — safe to modify, at the cost of extra allocation.
- Use `.base` and `.ctypes.data` to diagnose memory sharing; never rely on `id()`.

| Operation | View or Copy? |
|---|---|
| Basic slicing `arr[a:b]` | View |
| `arr.view()` | View |
| `reshape()` / `ravel()` | View *(usually)* |
| `transpose()` / `.T` | View |
| `arr.copy()` / `flatten()` | Copy |
| Fancy indexing | Copy |
| Boolean masking | Copy |
| Arithmetic `arr + n` | Copy |
