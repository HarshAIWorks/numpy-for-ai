1. np.random.seed()

```Produces the same random numbers every time.```<BR>This makes experiments reproducible.

<h3>5. np.random.choice()

Randomly ```selects values.```

<h3>6. np.random.shuffle()

```Shuffles the original array.```

````python
arr = np.array([1,2,3,4,5]) 
np.random.shuffle(arr) 
print(arr)
````

````python
arr
````

**Important**

```It modifies the original array.```

<h3>7. np.random.permutation()

Returns a ```shuffled copy.```

````python
arr = np.array([1,2,3,4,5]) 
new_arr = np.random.permutation(arr) 
print(arr) 
print(new_arr)
````

<h2>Common Mistakes

❌ Forgetting seed()

Results change every run.

❌ Assuming rand() returns integers

It always returns floating-point numbers between 0 and 1.


<h2>Best practices

<li>Use seed() for reproducible experiments.
<li>Use permutation() when you need to preserve the original data.
<li>Use shuffle() only when modifying the original array is acceptable.
<li>Use choice() for random sampling tasks.
