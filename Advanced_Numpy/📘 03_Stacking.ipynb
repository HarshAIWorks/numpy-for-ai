<h1>📘 03_Stacking.ipynb

Stacking means ```combining two or more arrays into a single array.```

<h3>Functions Covered

```
Function	               Purpose	                               AI Importance
np.concatenate()	       Join arrays along an existing axis	   ⭐⭐⭐⭐⭐
np.vstack()                Stack vertically (row-wise)	           ⭐⭐⭐⭐⭐
np.hstack()	               Stack horizontally (column-wise)	       ⭐⭐⭐⭐⭐
np.column_stack()	       Combine arrays as columns	           ⭐⭐⭐⭐☆
np.dstack()	               Stack along depth (3rd axis)	           ⭐⭐⭐☆☆

<h3>1. np.concatenate()

Joins arrays along an existing axis.

```
np.concatenate((arr1, arr2), axis=0)

<h5>1D Example

````python
import numpy as np

arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
result = np.concatenate((arr1, arr2))
print(result)
````

**2D Example**

Columns wise (axis=0)

````python
arr1 = np.array([[1,2],
                 [3,4]])

arr2 = np.array([[5,6],
                 [7,8]])

print(np.concatenate((arr1, arr2), axis=0))
````

Row wise(axis =1)

````python
print(np.concatenate((arr1, arr2),axis=1))
````

<h3>2. np.vstack()

Stacks arrays ```vertically (row-wise).```

````python
arr1 = np.array([[1,2],
                 [3,4]])

arr2 = np.array([[5,6],
                 [7,8]])
print(np.vstack((arr1, arr2)))
````

<h3>3. np.hstack()

````python
arr1 = np.array([[1,2],
                 [3,4]])

arr2 = np.array([[5,6],
                 [7,8]])
print(np.hstack((arr1, arr2)))
````

<h3>4. np.column_stack()

Treats ```each input array as a column.```

````python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print(np.column_stack((arr1, arr2)))
````

<h3>5. np.dstack()

Stacks arrays ```along the third dimension (depth).```

````python
arr1 = np.array([[1,2],
                 [3,4]])

arr2 = np.array([[5,6],
                 [7,8]])

print(np.dstack((arr1, arr2)))
````
