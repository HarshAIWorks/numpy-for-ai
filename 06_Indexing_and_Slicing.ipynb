## 🧩 What Is Indexing?

process of accessing a single element from a NumPy array using its position (index).<br>
Indexing in Python starts from 0, not 1.

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Why Is It Important?</span></div>

```
In AI, we constantly access:

A particular feature
A specific pixel
A token in a sentence
A row in a dataset
A value in a tensor

Without indexing, working with data would be impossible.
```

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Where Is It Used in AI?</span></div>

```
✅ Machine Learning datasets
✅ Deep Learning tensors
✅ Computer Vision images
✅ NLP token sequences
✅ LLM embeddings
✅ AI Agent memory
```

<div style="background:#1E2235; border-radius:10px; padding:14px 22px; margin:28px 0 14px 0; font-family:'Segoe UI',sans-serif; display:flex; align-items:center; gap:12px; border-left:5px solid #3D6FDB;">
<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">1D Array Indexing — arr[index]</code>
</div>

````python
import numpy as np
````

````python
array_1d=np.arange(10,60, step=10)
array_1d
````

```                                 
Index                                 Negative Indexing  
                                                         
 0   1  2  3  4                      -5  -4  -3 - 2  -1  
 ↓   ↓  ↓  ↓  ↓                       ↓   ↓   ↓   ↓   ↓  
[10 20 30 40 50]                     [10  20  30  40  50]
```

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ First and Last Elements</span></div>

````python
first_element= array_1d[0] #0r first_element= array_1d[-5]
last_element= array_1d[-1] #0r last_element= array_1d[4]
print("first: ",first_element)
print("last_element: ",last_element)
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Other Elements — nth element has index n-1</span></div>

````python
second_element= array_1d[1] #0r second_element= array_1d[-4]
third_element= array_1d[2]  #0r third_element= array_1d[-3]
print("second_element: ",second_element)
print("third_element: ",third_element)
````

<div style="border:2px solid #F59E0B; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#D97706; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">⚠️</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">Warning — out-of-range index</span></div>
<div style="background:#FFFBEB; padding:14px 18px; color:#78350F; font-size:0.97em; line-height:1.6;">
Accessing an index beyond the array's length raises <code>IndexError</code>. The cell below intentionally demonstrates this — a teaching example.
</div>
</div>

````python
array_1d[5]
````

Because the valid indices are: 0 1 2 3 4

<div style="background:#1E2235; border-radius:10px; padding:14px 22px; margin:28px 0 14px 0; font-family:'Segoe UI',sans-serif; display:flex; align-items:center; gap:12px; border-left:5px solid #3D6FDB;"><h3>
<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">2D Array Indexing — arr[row, column]</code>
</div>

````python
array_2d=np.array([[10,20,30],[40,50,60],[70,80,90]])
array_2d
````

```
         Column       
        0   1   2     
      ┌─────────────┐ 
Row 0 │10  20  30 │   
Row 1 │40  50  60 │   
Row 2 │70  80  90 │   
      └─────────────┘ 


<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ First Element and Last Element</span></div>

````python
first=array_2d[0,0]
last=array_2d[-1,-1]

print("First element: ",first)
print("Last element: ",last)
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Accessing Entire Rows ⭐⭐⭐⭐⭐arr[row] (Most Importanat)</span></div>

````python
row1=array_2d[0] #first row
row2=array_2d[1] #second row

print("row 1: ",row1)
print("row 2: ",row2)
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Accessing Entire Columns ⭐⭐⭐⭐⭐ — arr[:, col] (most important)</span></div>

````python
col1= array_2d[:,0] #you will learn in slicing about : this [:] - it means something all 
col_last=array_2d[:,-1]

print("First colums: ",col1)
print("Last colums: ",col_last)
````

<div style="border:2px solid #2563EB; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#1D4ED8; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">💡</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">Tip</span></div>
<div style="background:#EFF6FF; padding:14px 18px; color:#1E3A5F; font-size:0.97em; line-height:1.6;">
The <code>:</code> symbol means "all elements along this axis" — you will use it constantly once you reach slicing.
</div>
</div>

<div style="background:#1E2235; border-radius:10px; padding:14px 22px; margin:28px 0 14px 0; font-family:'Segoe UI',sans-serif; display:flex; align-items:center; gap:12px; border-left:5px solid #3D6FDB;"><h3>
<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">3D Array Indexing — arr[layer, row, column]</code>
</div>

````python
array_3d= np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
array_3d
````

```
Layer 0

      col0  col1
row0   1     2
row1   3     4

Layer 1

      col0 col1
row0   5    6
row2   7    8


<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ First and Last Element</span></div>

````python
first=array_3d[0,0,0]
array_3d.shape #(2,2,2) 2 layers, 2 rows, 2 colums.
               # last element: layer 2(index1), row2(index1), col2(index1) ***always index==position-1
last=array_3d[1,1,1]
print("First element: ",first)
print("Last element: ",last)
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Specific Value</span></div>

Value 4 and Value 6

````python
# 4 is at layer1 and 2 col, 2 row  *** dont prefer above image of row,col,layer- its already in index form
print(array_3d[0,1,1])
#6 is at layer 2, row 1 and second col
print(array_3d[1,0,1])
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ ❌ Forgetting That Indexing Starts at 0</span></div>

<div style="border:2px solid #F59E0B; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">

<span style="color:#E8A84A; font-size:1.15em; font-weight:800;">PART 2</span>
<span style="color:#FFFFFF; font-weight:700; font-size:1.15em;">Slicing</span>
</div>

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ What Is Slicing?</span></div>

Process of extracting **multiple elements** (a portion of an array) from an existing array.

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Why Is It Important?</span></div>

In AI, we rarely work with an entire dataset at once.<br>

We often need to:<br>

Select specific rows<br>
Select specific columns<br>
Crop images<br>
Extract feature columns<br>
Create training and testing datasets<br>
Process batches of data<br>

All of these use slicing.

**Slicing Syntax:**
```
array[start : stop : step]
```

<div style="border:2px solid #2563EB; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#1D4ED8; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">💡</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">Golden Rule</span></div>
<div style="background:#EFF6FF; padding:14px 18px; color:#1E3A5F; font-size:0.97em; line-height:1.6;">
<b>Start</b> → Included ✅ &nbsp;&nbsp;|&nbsp;&nbsp; <b>Stop</b> → Excluded ❌
</div>
</div>

<div style="background:#1E2235; border-radius:10px; padding:14px 22px; margin:28px 0 14px 0; font-family:'Segoe UI',sans-serif; display:flex; align-items:center; gap:12px; border-left:5px solid #3D6FDB;"><h3>
<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">1D Array Slicing</code>
</div>

````python
import numpy as np
arr= np.arange(10,70,10)
arr
````

```
0   1   2   3   4   5 : Index
↓   ↓   ↓   ↓   ↓   ↓

10  20  30  40  50  60 : items

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 1. Basic Slicing</span></div>

````python
arr[1:4] #include index 1,2,3 {4; will excluded}
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 2. From Beginning</span></div>

````python
arr[:4]
````

Since start is omitted: <br>```Start = 0```

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 3. Till End</span></div>

````python
arr[2:]
````

Since stop is omitted:<br>```Go until the last element.```

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 4. Entire Array</span></div>

````python
arr[:]
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 5. Using Step</span></div>

````python
arr[::2]
````

```Take every 2nd element.```<br><br>Selected indices:
```
   0  
   2  
   4  

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 6. Reverse Array ⭐⭐⭐⭐⭐ (Important) </span></div>

````python
arr[::-1]
````

```
Step = -1
```
Move backwards.

Reverse every second element:

````python
arr[::-2]
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 7. Negative Slicing</span></div>

```
-6 -5 -4 -3 -2 -1

10 20 30 40 50 60

````python
arr[-4:-1]
````

<div style="background:#1E2235; border-radius:10px; padding:14px 22px; margin:28px 0 14px 0; font-family:'Segoe UI',sans-serif; display:flex; align-items:center; gap:12px; border-left:5px solid #3D6FDB;"><h3>
<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">2D Array Slicing — arr[row_slice, column_slice]</code>
</div>

````python
arr= np.arange(10,100,10).reshape(3,3)
arr
````

```
          Column       

         0   1   2
       ┌────────────┐

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 1. Rows Only ⭐⭐⭐⭐⭐ (Important) </span></div>

— First Two Rows

````python
arr[:2]
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 2. Columns Only ⭐⭐⭐⭐⭐ (Important)</span></div>

First Column

````python
arr[:,0]
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 3. With Steps</span></div>

a) First and last columns

````python
arr[:,0:3:2]
````

b) First and third row

````python
arr[0:3:2]
````

c) Every second row

````python
arr[::2]
````

d) Reverse columns

````python
arr[:,::-1]
````

e) Reverse entire matrix

````python
arr[::-1,::-1]
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 4. Matrix Selection — Bottom 2×2 Matrix</span></div>

```
          Column       

         0   1   2
       ┌────────────┐
row 0    │        │

       └────────────┘`

````python
arr[1:3,1:3]
````

<div style="background:#1E2235; border-radius:10px; padding:14px 22px; margin:28px 0 14px 0; font-family:'Segoe UI',sans-serif; display:flex; align-items:center; gap:12px; border-left:5px solid #3D6FDB;"><h3>
<code style="background:#2A2F47; color:#E8A84A; font-size:1.1em; font-weight:700; padding:5px 14px; border-radius:6px; font-family:'Courier New',monospace;">3D Array Slicing — arr[layer_slice, row_slice, column_slice]</code>
</div>

<div style="border:2px solid #16A34A; border-radius:10px; margin:18px 0; overflow:hidden; font-family:'Segoe UI',sans-serif; box-shadow:0 2px 8px rgba(0,0,0,0.08);">
<div style="background:#15803D; padding:8px 16px; display:flex; align-items:center; gap:8px;"><span style="font-size:1.1em;">ℹ️</span><span style="color:#FFFFFF; font-weight:700; font-size:0.95em; letter-spacing:0.5px;">Info</span></div>
<div style="background:#F0FDF4; padding:14px 18px; color:#14532D; font-size:0.97em; line-height:1.6;">
Earlier, <code>[row_slicing]</code> alone gave you rows — but for a 3D array, that same first position now controls <b>layer</b> slicing instead.
</div>
</div>

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ Defining a 3D Array</span></div>

````python
arr=np.arange(1,28).reshape(3,3,3)
arr
````

<div style="background:#1E3A5F; border-radius:6px; padding:7px 14px; margin:14px 0 6px 0; display:inline-block; font-family:'Segoe UI',sans-serif;"><span style="color:#93C5FD; font-weight:700; font-size:0.97em;">▸ 1. Layers Only — First Layer</span></div>

````python
arr[0]
````

<a id="best-practices"></a>

## ✅ Best Practices

- Use **slicing** instead of loops whenever possible — it is faster and more readable.
- Use **negative slicing** to access elements from the end without knowing the array's length.
- Learn these patterns by heart:

```
arr[:]         # entire array
arr[:5]        # first 5 elements
arr[5:]        # everything from index 5 onward
arr[::2]       # every second element
arr[::-1]      # reversed array
arr[:,0]       # first column (2D)
arr[:,:]       # entire 2D array
arr[:,::-1]    # reverse columns (2D)
```

<a id="interview-questions"></a>

## 🎤 Interview Questions

---

### 🟢 Beginner

**Q1.** What is the difference between indexing and slicing?
<details><summary>▶ Answer</summary>
Indexing accesses a <b>single element</b> and returns a scalar. Slicing extracts a <b>range of elements</b> and always returns an array (even if that array has only one element).
</details>

**Q2.** What does `arr[-1]` return?
<details><summary>▶ Answer</summary>
The <b>last</b> element of the array — negative indices count backwards from the end, starting at -1.
</details>

**Q3.** In `arr[2:5]`, which indices are included?
<details><summary>▶ Answer</summary>
Indices 2, 3, and 4. The stop index (5) is always <b>excluded</b>.
</details>

---

### 🟡 Intermediate

**Q4.** How do you select an entire column from a 2D array?
<details><summary>▶ Answer</summary>
<code>arr[:, col_index]</code> — the <code>:</code> in the row position means "all rows," combined with a single column index.
</details>

**Q5.** What does `arr[::-1]` do, and how is it different from `arr[::-2]`?
<details><summary>▶ Answer</summary>
<code>arr[::-1]</code> reverses the entire array (step -1, moving backwards one at a time). <code>arr[::-2]</code> reverses it while also skipping every other element (step -2).
</details>

**Q6.** For a 3D array, what does the first index position control?
<details><summary>▶ Answer</summary>
The <b>layer</b> (the outermost axis) — unlike a 2D array where the first position controls the row.
</details>

---

### 🔴 Advanced

**Q7.** What is the difference between `arr[2]` and `arr[2:3]` for a 1D array?
<details><summary>▶ Answer</summary>
<code>arr[2]</code> returns a <b>scalar</b> (the single value at index 2). <code>arr[2:3]</code> returns a <b>1-element array</b> — same value, different type/shape.
</details>

**Q8.** How would you select every second column from every layer of a 3D array?
<details><summary>▶ Answer</summary>
<code>arr[:, :, 0::2]</code> — all layers, all rows, then step through columns by 2.
</details>

**Q9.** Why does `arr[1:4]` on `np.arange(5)` return only 3 elements instead of 4?
<details><summary>▶ Answer</summary>
Because the stop index is exclusive: it returns indices 1, 2, 3 — three elements — not index 4.
</details>

### Function Quick Reference

| Pattern | Meaning |
|---|---|
| `arr[i]` | Single element at index i |
| `arr[i:j]` | Elements from i (included) to j (excluded) |
| `arr[::k]` | Every kth element |
| `arr[::-1]` | Reversed array |
| `arr[:, c]` | Entire column c (2D) |
| `arr[r, :]` | Entire row r (2D) |
| `arr[l, :, :]` | Entire layer l (3D) |
